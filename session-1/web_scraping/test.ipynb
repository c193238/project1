```python
from requests_html import HTMLSession

url = "https://businesspostbd.com/economy/uae-eager-to-enhance-collab-with-bangladesh"

session = HTMLSession()
response = session.get(url)
```

```python
title_tag = response.html.find('h1')
print("Title: ", title_tag[0].text)
author_tag = response.html.find('div.w3-col > strong')
print("Author: ", author_tag[0].text)
time_tag = response.html.find('div.w3-col > span')
print("Time: ", time_tag[0].text)
category = response.html.find('p.w3-text-gray > a')
print("Category: ", category[1].text)
image = response.html.find('div.w3-margin-bottom > img')
print("Image : ", image[0].attrs["src"])
target_div = response.html.find('div.ub-margin-bottom-32.w3-justify', first=True)

# Get all <p> tags within this div
paragraphs = target_div.find('p')

# Extract and concatenate the text from all paragraphs
print("Body: ")
body = ''
for p in paragraphs:
    print(p.text)
    body += p.text

  
```
